```python
#Criando nosso Data/frame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a Receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo Total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Qual o custo Total?
round(df["custo"].sum(), 2)
```

```python
#Qual o lucro total?
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["lucro"].sum(), 2)
```

```python
#Criando uma coluna com o total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

### **Agora, queremos saber a média de tempo de envio para cada Marca**

```python
#Transformando a diferencia de dias para o envio em um valor numérico
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo de coluna para Tempo_envio
df.dtypes
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

### **Missing Values**

```python
#Verificando se há dados faltantes
df.isnull().sum()
```

**E, se quisermos saber o Lucro por Ano e por Marca**

```python
#Agrupamento por Ano e por Marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#Tirando a notação científica da visualização
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("produto");
```

```python
#Lucro por ano
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Gráfico de linhas com o lucro por mês no ano de 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Gráfico do lucro por marca
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro");
```

```python
#Gráfico do lucro por classe
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
